```python
summary = []

for table_name, df in all_tables.items():
    
    summary.append({
        'Table Name': table_name,
        'Rows': df.shape[0],
        'Columns': df.shape[1],
        'Missing Values': df.isnull().sum().sum()
    })

summary_df = pd.DataFrame(summary)

print(summary_df)
```

```python
# finding missing values 

for table_name, df in all_tables.items():
    
    print(f"\n===== {table_name} =====")
    
    blank_values = ((df == ' ') | (df == '')).sum()
    
    print(blank_values)
    
```

```python
# finding  missing values in another method
import numpy as np

for table_name, df in all_tables.items():
    
    df.replace(
        [' ', '', 'NA', 'N/A', 'null', 'None'],
        np.nan,
        inplace=True
    )
    
    print(f"\n===== {table_name} =====")
    
    print(df.isnull().sum())
```

```python
# replace missing values with zeros for a specific table

df = all_tables['orders']

print(df)

if 'days_since_prior_order' in df.columns:
    
    df['days_since_prior_order'] = (df['days_since_prior_order'].fillna(0))
    
print("missing values handle")

```
